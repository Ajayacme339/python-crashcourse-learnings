Returning a Dictionary

```python
def build_person(first_name,last_name):
    """Build a dictionary of user information."""
    person = {'first':first_name ,'last':last_name}
    return person

person_name = build_person('Ajay','Kumar')
print(person_name)    
```

```python
def build_person(first_name,last_name,age=''):
    """Build a dictionary of user information."""
    person = {'first':first_name ,'last':last_name}
    if age:
        person['age'] = age
    return person

person_name = build_person('Ajay','Kumar',32)
print(person_name)
```

```python
def get_formatted_name(first_name, last_name):
    """Return a full name, neatly formatted."""
    full_name = first_name + ' ' + last_name
    return full_name.title()
while True:
    print("\nPlease tell me your name:")
    print("(enter 'q' at any time to quit)")
    f_name = input("First name: ")
    if f_name == 'q':
        break
    l_name = input("Last name: ")
    if l_name == 'q':
        break
        
formatted_name = get_formatted_name(f_name, l_name)
print("\nHello, " + formatted_name + "!")
```

Mixing Positional and Arbitrary Arguments

```python
def make_pizza(size,*toppings):
    """summarize pizza we are making"""
    print("\nMaking a " + str(size) +
          "-inch with  following toppings:")
    for topping in toppings:
        print("-" + topping)
    
make_pizza(16,'cheese')
make_pizza(12,'cheese','garlic')
#make_pizza('cheese','peoprini')
```

```python
def build_profile(first, last, **user_info):
    """Build a dictionary containing everything we know about a user."""
    profile = {}
    profile['first_name'] = first
    profile['last_name'] = last
    for key ,value in user_info.items():
        profile[key] = value
    return profile

user_profile = build_profile('Ajay','Kumar',age=32)
print(user_profile)
```

```python
user_profile = build_profile(age=32,'Ajay','Kumar')
print(user_profile)
```

```python
user_profile = build_profile(age=32,first='Ajay',last='Kumar')
print(user_profile)
```
