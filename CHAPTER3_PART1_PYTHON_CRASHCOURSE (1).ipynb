 •Use sorted() to print your list in alphabetical order without modifying the 
actual list .
 •Show that your list is still in its original order by printing it .
 •Use sorted() to print your list in reverse alphabetical order without chang
ing the order of the original list .
 •Show that your list is still in its original order by printing it again .
 •Use reverse() to change the order of your list . Print the list to show that its 
order has changed .
 •Use reverse() to change the order of your list again . Print the list to show 
it’s back to its original order .
 •Use sort() to change your list so it’s stored in alphabetical order . Print the 
list to show that its order has been changed .
 •Use sort() to change your list so it’s stored in reverse alphabetical order . 
Print the list to show that its order has changed. 

```python
dream_place = ['lasvegas','france','japan','switzerland','germany']
print("Here is the original order of list:")
print(dream_place)
print("\nPrinting list in alphabetical order using sorted function:")
print(sorted(dream_place))
print("\nMy list is still in the original order:")
print(dream_place)
print("\nPrinting list in reverse alphabetical order using sorted function:")
print(sorted(dream_place,reverse = True))
print("\nMy list is still in the original order:")
print(dream_place)
dream_place.reverse()
print("\nMy list is still in the reversed order:")
print(dream_place)
dream_place.reverse()
print("\nMy list is back in the original order:")
print(dream_place)
print("\nMy list is in the sorted order:")
dream_place.sort()
print(dream_place)
print("\nMy list is in the  reverse sorted order:")
dream_place.sort(reverse=True)
print(dream_place)
```

3-10.Every Function: Think of something you could store in a list . For example, 
you could make a list of mountains, rivers, countries, cities, languages, or any
thing else you’d like . Write a program that creates a list containing these items 
and then uses each function introduced in this chapter at least once.

```python
cities = ['sitamarhi','aurai','runnisaidpur','patna','bokaro']
print("Here is the Original City list:")
print(cities)
print("\nCity list in the sorted order:")
sorted(cities)
print(sorted(cities))
print("\nHere is the Original City list:")
print(cities)
print("\nCity list in the reverse sorted order:")
print(sorted(cities,reverse=True))
print("\nHere is the Original City list again:")
print(cities)
cities.sort()
print("\nHere is the city list in the sorted order:")
print(cities)
print("\nHere is the Original City list again:")
print(cities)
cities.sort(reverse=True)
print("\nHere is the city list in the reverse sorted order:")
print(cities)
print("\nThe number of city in the city list is:",len(cities))
cities.insert(0,'chennai')
cities.append('pune')
print("\nHere is the city list after adding cities in list:")
print(cities)
popped_cities = cities.pop()
print("\nHere is the city list after popping cities in list:")
print(cities)
print("\nHere is the popped city list:",popped_cities)
print(cities)
del cities[0]
print("\nHere is the city list after deleting a city in list:")
print(cities)
print("\nHere is the city list in a reversed order:")
cities.reverse()
print(cities)
print("\nHere is the city list after removing a city in list:")
cities.remove('patna')
print(cities)
print("\nHere is the modified city list:")
cities[1] = 'Amelia'
print(cities)
```
